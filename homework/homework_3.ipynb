```python
import pandas as pd
import numpy as np
```

<b>Dataset</b>

```python
data = "https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv"
df = pd.read_csv(data)
df
```

<b>Features</b>

```python
features = ['Make', 'Model', 'Year', 'Engine HP', 'Engine Cylinders', 'Transmission Type', 
            'Vehicle Style', 'highway MPG', 'city mpg', 'MSRP']
```

<b>Data preparation</b>

```python
df = df[features]
df.columns = df.columns.str.replace(' ', '_').str.lower()
categorical_columns = list(df.dtypes[df.dtypes == 'object'].index)
for column in categorical_columns:
    df[column] = df[column].str.replace(' ', '_').str.lower()
df
```

```python
df = df.fillna(0)
df.isnull().sum()
```

```python
df['price'] = df['msrp']
del df['msrp']
df
```

```python
df.transmission_type.value_counts()
```

<b>Question 2</b>

```python
numerical = ['year', 'engine_hp', 'engine_cylinders', 'highway_mpg', 'city_mpg']
categorical = ['make', 'model', 'transmission_type', 'vehicle_style']
```

```python
for variable in numerical:
    print(variable)
    print(df[numerical].corrwith(df[variable]))
    print()
```

<b>Making "price" binary</b>

```python
mean_price = df.price.mean()
mean_price
```

```python
df['above_average'] = (df.price > mean_price).astype(int)
price = df['price']
df
```

<b>Splitting the data</b>

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25)
len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train['above_average']
y_val = df_val['above_average']
y_test = df_test['above_average']
y_train_price = np.log1p(df_train['price'].values)
y_val_price = np.log1p(df_val['price'].values)
y_test_price = np.log1p(df_test['price'].values)
```

```python
del df_train['above_average']
del df_val['above_average']
del df_test['above_average']

del df_train['price']
del df_val['price']
del df_test['price']
```

<b>Question 3</b>

```python
from sklearn.metrics import mutual_info_score
```

```python
def mutual_info_above_average_score(series):
    return mutual_info_score(series, y_train)
```

```python
mutual_info = df_train[categorical].apply(mutual_info_above_average_score)
mutual_info.sort_values(ascending=True).to_frame('Mutual Information')
```

<b>Question 4</b>

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
train_dicts = df_train[categorical + numerical].to_dict(orient='records')
val_dicts = df_val[categorical + numerical].to_dict(orient='records')
dv = DictVectorizer(sparse=False)
```

```python
X_train = dv.fit_transform(train_dicts)
X_val = dv.transform(val_dicts)
```

```python
model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_val)
accuracy = (y_pred == y_val).mean()
round(accuracy, 2)
```

<b>Question 5</b>

```python
features_to_eliminiate = ['transmission_type', 'year', 'engine_hp', 'city_mpg']

for feature in features_to_eliminiate:
    categorical = ['make', 'model', 'transmission_type', 'vehicle_style'] 
    numerical = ['year', 'engine_hp', 'engine_cylinders', 'highway_mpg', 'city_mpg'] 
    features = categorical + numerical
    
    features.remove(feature)
    
    train_dicts = df_train[features].to_dict(orient='records')
    val_dicts = df_val[features].to_dict(orient='records')
    dv = DictVectorizer(sparse=False)
    X_train = dv.fit_transform(train_dicts)
    X_val = dv.transform(val_dicts)

    model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    accuracy_difference = accuracy - (y_pred == y_val).mean()
    
    print(feature, accuracy_difference)
```

<b>Question 6</b>

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
alpha_values = [0, 0.01, 0.1, 1, 10]
```

```python
for value in alpha_values:
    train_dicts = df_train[categorical + numerical].to_dict(orient='records')
    val_dicts = df_val[categorical + numerical].to_dict(orient='records')
    dv = DictVectorizer(sparse=True)
    
    X_train = dv.fit_transform(train_dicts)
    X_val = dv.transform(val_dicts)
    
    model = Ridge(alpha=value, solver='sag', random_state=42)
    model.fit(X_train, y_train_price)
    y_pred = model.predict(X_val)
    
    score = mean_squared_error(y_val_price, y_pred)
    print(f'For alpha = {value} the score is {round(score, 3)}')
```
